```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data = {
  "Classification": ['Logistic Regression', 'Support Vector Machine', 'Decision Tree','K Nearest Neighbors'],
  "Accuracy": [0.8464285714285713, 0.8482142857142856, 0.8767857142857143,0.8482142857142858]
}
df = pd.DataFrame(data)
df.set_index('Classification',inplace=True)
df
```

```python
df.plot(kind='bar', figsize=(5, 3))
plt.xlabel('Classification') # add to x-label to the plot
plt.ylabel('Accuracy') # add y-label to the plot
plt.title('Accuracy per Classification Method') # add title to the plot
plt.legend(bbox_to_anchor=(0.99, -0.7), loc='upper right')
plt.show()
```
